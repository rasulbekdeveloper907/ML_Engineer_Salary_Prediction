# Kategorik ustunlarni Label Encoding qilish

```python
from sklearn.preprocessing import LabelEncoder  # sklearn’dan LabelEncoder chaqiramiz

# DataFrame’dagi faqat matnli (object) ustunlarni aniqlaymiz
cat_cols = df.select_dtypes(include="object").columns

# Har bir ustun uchun encoderlarni keyin ishlatish uchun saqlash
label_encoders = {}

# Kategorik ustunlarni sonlarga aylantiramiz
for col in cat_cols:
    le = LabelEncoder()         # Encoder obyektini yaratamiz
    df[col] = le.fit_transform(df[col])  # Har bir kategoriyani raqamga o‘tkazamiz
    label_encoders[col] = le    # Keyin kerak bo‘lsa qayta ishlatish uchun saqlaymiz

print("Kategorik ustunlar kodlandi ✅")
```

## 📝 Umumiy xulosa

1. **Ko‘plab ustunlar kategorik bo‘lib, son sifatida qaralgani uchun skewness yuqori**.  
2. **salary va salary_in_usd** ustunlariga **log transformatsiya** qo‘llash foydali.  
3. **employment_type, job_title, salary_currency, company_location, employee_residence, company_size** ustunlari uchun **One-Hot Encoding yoki Target Encoding** tavsiya qilinadi.  
4. **is_remote va is_large_company** ustunlari **imbalanced (nomutanosib)** bo‘lgani uchun balanslash texnikalari (**SMOTE, oversampling, undersampling**) kerak bo‘lishi mumkin.


```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, PowerTransformer

# ================================
# 1. Salary ustunlarini log transformatsiya qilish
# ================================
df["salary"] = np.log1p(df["salary"])                 # log(1+x) -> outlierlarni kamaytiradi
df["salary_in_usd"] = np.log1p(df["salary_in_usd"])
df["salary_per_hour"] = np.log1p(df["salary_per_hour"])

# ================================
# 2. Kategorik ustunlarni encoding qilish
# ================================
categorical_cols = ["employment_type", "job_title", "salary_currency",
                    "employee_residence", "company_location", "company_size"]

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])   # LabelEncoder orqali raqamlashtiramiz

# ================================
# 3. work_year va experience_level ustunlari (Yeo-Johnson)
# ================================
pt = PowerTransformer(method="yeo-johnson")

df[["work_year", "experience_level"]] = pt.fit_transform(
    df[["work_year", "experience_level"]]
)

# ================================
# 4. is_remote, is_large_company
# ================================
# Bu ustunlar binary (0/1) bo‘lgani uchun transformatsiya qilinmaydi

# ================================
# ✅ Natijani ko‘rish
# ================================
print(df.head())
```
