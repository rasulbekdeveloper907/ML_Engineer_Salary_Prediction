```python
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\Rasulbek_Ruzmetov\Desktop\SML_R_Project\Data\Raw_Data\ML_Engineeer.csv")
```

```python
df.head()
```

```python
# Agar datasetda "salary_in_usd" ustuni bo‘lsa, soatbay oylikni hisoblaymiz
if "salary_in_usd" in df.columns:
    df["salary_per_hour"] = df["salary_in_usd"] / (40 * 52)  # 40 soat/hafta * 52 hafta/yil

# Agar "remote_ratio" ustuni bo‘lsa, to‘liq remote ishlovchilarni belgilaymiz
if "remote_ratio" in df.columns:
    df["is_remote"] = (df["remote_ratio"] == 100).astype(int)  # 100% remote = 1, boshqasi = 0

# Agar "company_size" ustuni bo‘lsa, yirik kompaniyalarni belgilaymiz
if "company_size" in df.columns:
    df["is_large_company"] = (df["company_size"] == "L").astype(int)  # L = Large => 1

print("Yangi xususiyatlar qo‘shildi ✅")
```
